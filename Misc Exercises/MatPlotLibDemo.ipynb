```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
```

```python
plt.plot(x,y, 'r--')
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

# Iterating
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

# Indexing
axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig1 = plt.figure(figsize=(8,2))

ax = fig1.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig2, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig2.savefig('my_picture.png', dpi=200)
```

```python
fig3 = plt.figure()

ax = fig3.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.plot(y,x)
```

```python
fig4 = plt.figure()

ax4 = fig4.add_axes([0,0,1,1])
ax4.plot(x,x**2, label ="X Squared")
ax4.plot(x,x**3, label = 'X cubed')

ax4.legend(loc=2)
```

```python
fig5 = plt.figure()
ax = fig5.add_axes([0,0,1,1])

# Changing the display
ax.plot(x,y,color='purple', linewidth=4, alpha=0.5, linestyle='-.' )
```

```python
fig6 = plt.figure()
ax = fig6.add_axes([0,0,1,1])

# Changing the display
ax.plot(x,y,color='red', linewidth=5, linestyle='-', marker="o", markersize=14,
       markerfacecolor="blue")
```
