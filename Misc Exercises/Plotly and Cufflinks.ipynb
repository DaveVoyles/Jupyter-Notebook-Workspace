```python
# DATA
df = pd.DataFrame(np.random.randn(100,4), columns= 'A B C D'.split())
```

```python
df.head()
```

```python
#Random values
df2 = pd.DataFrame({'Category':['A', 'B', 'C'], "Values": [32,43,50]})
```

```python
df2
```

```python
df.plot()
```

```python
# Same plot, but interactive w/ plotly
df.iplot()
```

```python
df.iplot(kind='scatter', x="A", y="B", mode="markers")
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
# 3D data
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,1], 'z':[500,400,300,200,100]})
```

```python
df3
```

```python
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
# Useful to display data such as stock
# Shows the spread arrays against eachother
df[['A', 'B']].iplot(kind='spread')
```

```python
# Similar to scatter plot, but will change size based on another column value
# Uses: Word GDP, population, happines factor, etc
df.iplot(kind='bubble', x='A', y='B', size='C' )
```
