### Fill in missing age values
Use avg age values to impute age based off of Pclass

```python
def impute_age(cols):
    Age    = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37 # Avg age in Pclass 1
        elif Pclass == 2:
            return 29 # Avg age in Pclass 2
        else: 
            return 24 # Avg age in Pclass 3
        
    else:
            return Age #Already know their age, so just return that   
```

```python
# Grab Age & Pclass columns, and apply the avg age function to them. Set this value to the age column.
train['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
# All of the missing variables for the age column are now filled.
sns.heatmap(train.isnull()     , 
            yticklabels = False,
            cbar        = False,
            cmap        = 'viridis')
```

### Convert categorical variables into integer vars

```python
# PROBLEM: One column is a perfect predictor of the other column.'Multi-colineariaty'
pd.get_dummies(train['Sex'])
```

```python
# Drops the first column, & we only need to know if a user was male or not.
# Feed this into ML model
sex = pd.get_dummies(train['Sex'], drop_first=True)
```

```python
embark = pd.get_dummies(train['Embarked'], drop_first = True)
```

```python
embark.head()
```

### Take new columns, add them to our data frame

```python
# Set axis=1 to add them as -new- columns
train = pd.concat([train, sex, embark], axis=1)
```

```python
train.head(2)
```

### Remove columns which are not relevent to the model

```python
train.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
train.head()
```

### Split data into dest & train

```python
X = train.drop('Survived', axis=1)
y = train['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
# Make class predictions for the testing set
predictions = logmodel.predict(X_test)
```

# Evaluating the model
[source](https://www.youtube.com/watch?v=85dtiMz9tSo)

### Predict class probabilities

```python
probability = logmodel.predict_proba(X_test)
print(probability)
```

### Eval Metrics

```python
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix
```

#### Accuracy tells us what percentage of predictions in our y_test class were correct

```python
print ("ACCURACY SCORE:", accuracy_score(y_test, predictions))
print ("ROC_AUC_SCORE: ", roc_auc_score(y_test, probability[:, 1]))
# evaluate the accuracy of a classification
print("\n CONFUSION MATRIX: \n", confusion_matrix(y_test, predictions))

print('\n    ACTUAL:' , y_test.values[0:25])
print(' PREDICTED:', predictions[0:25])
```

```python
confusion = confusion_matrix(y_test, predictions)
TP        = confusion[1,1]
TN        = confusion[0,0]
FP        = confusion[0,1]
FN        = confusion[1,0]
```

**Sensitivity:** When the actual value is *positive*, how often is the prediction correct? (True positive rate, recall)

**Specificity:** When the actual value is *negative*, how often is the prediction correct?

**FP Rate:**     When the actual value is *negative*, how often is the prediction incorrect?

**Precision:**   When a positive value is predictied, how often is the prediction correct?

```python
print ("TP RATE/SENSITIVITY/RECALL:",  TP / float(TP+FN))
print("SPECIFICITY                :" , TN / float(TN+FP))
print("FP RATE                    :" , FP / float(TN+FP))
print("PRECISION                  :" , TP / float(TP+FP))
```

### Classification report

```python
from sklearn.metrics import classification_report
```

```python
class_report = classification_report(y_test, predictions)
```

```python
print(class_report)
```

```python
# Source: https://stackoverflow.com/questions/28200786/how-to-plot-scikit-learn-classification-report
def plot_classification_report(cr, title='Classification report ', with_avg_total=False, cmap=plt.cm.Blues):

    lines = cr.split('\n')

    classes = []
    plotMat = []
    for line in lines[2 : (len(lines) - 3)]:
        #print(line)
        t = line.split()
        # print(t)
        classes.append(t[0])
        v = [float(x) for x in t[1: len(t) - 1]]
        print(v)
        plotMat.append(v)

    if with_avg_total:
        aveTotal = lines[len(lines) - 1].split()
        classes.append('avg/total')
        vAveTotal = [float(x) for x in t[1:len(aveTotal) - 1]]
        plotMat.append(vAveTotal)


    plt.imshow(plotMat, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    x_tick_marks = np.arange(3)
    y_tick_marks = np.arange(len(classes))
    plt.xticks(x_tick_marks, ['precision', 'recall', 'f1-score'], rotation=45)
    plt.yticks(y_tick_marks, classes)
    plt.tight_layout()
    plt.ylabel('Classes')
    plt.xlabel('Measures')
```

```python
plot_classification_report(class_report)
```
