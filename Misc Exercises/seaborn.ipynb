```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=50)
```

```python
# Compare two indices as x/y values
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# Compare two indices as x/y values
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# Compare two indices as x/y values
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex') # Pass in a column name of a category. Will color data points based on colum
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') 
```

```python
sns.rugplot(tips['total_bill']) # Draws a a dashed mark for every point in the data
```

```python
sns.distplot(tips['total_bill'], kde=False)
# Histogram has bins, whereas rugplot does not
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
import numpy as np
```

```python
# Use numpy estimator to determine standard deviation for a given category
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# Estimator counts the number of occurences (like .counts in Pandas)
sns.countplot(x='sex', data=tips)
```

```python
# Used to show the distribution of categorical data.
# Points outside of the 'whiskers' are outliers
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
# Split by per day AND smoker status
# Compare day-by-day behavior, and also if a person is a smoker.
# Hue gives a 3rd layer of data quickly and smoothly
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
# Like a box plot, but plots all  components that correspond to actual data points.
# Harder to read than a boxplot, though.
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
# Combo of strip plot and violin plot. Same distribution but showing all points 
# CON: Does not scale well w/ a large dataset.
sns.swarmplot(x='day', y='total_bill', data=tips)
```
