```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
# Show how values are relative to one another
tc = tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True)
```

```python
sns.heatmap(tc, annot=True, cmap='cool')
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
# Need to get this into -- Matrix Form --
flights
```

```python
# Pivot aligns data correctly 
flights.pivot_table(index='month', columns='year', values="passengers")
```

```python
fp = flights.pivot_table(index='month', columns='year', values="passengers")
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, linecolor='white', linewidths=1)
```

```python
# 2nd matrix-type plot in SNS
sns.clustermap(fp)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# Shows that sex has a similar linear fit, in terms of tip
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
# Separate by columns instead of color
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```
