```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

# Large
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

# Insert
ax2.plot(x,y)
ax2.set_title("zoom")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig, axes = plt.subplots(1,2, figsize=(12, 2))

axes[0].plot(x,y, color='blue', lw=3, ls='--')
axes[1].plot(x,z, color='red', lw=3)
```
